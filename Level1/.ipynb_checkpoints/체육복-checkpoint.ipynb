```python
def solution(n, lost, reserve):
    losts, reserves = sorted(list(set(lost) - set(reserve))), sorted(list(set(reserve) - set(lost)))
    temp = 0
    for i in losts:
        if i-1 in reserves:
            reserves.remove(i-1)
        elif i+1 in reserves:
            reserves.remove(i+1)
        else:
            temp += 1
    return n - temp
```

```python
print(solution(5, [2, 4], [1, 3, 5])) #5
print(solution(5, [2, 4], [3])) # 4
print(solution(3, [3], [1])) # 2
print(solution(10, [3,9,10], [3,8,9])) # 9
print(solution(3, [1,2,3], [1,2,3])) # 3
print(solution(11, [2, 4, 5, 8], [3, 5, 6])) # 9
```

# 해설
자기의 체육복을 잃어버렸을 경우 남에게 빌릴 수 있어도 자신이 사용하는 것이 최우선적이므로 losts, reserves에 겹치는 부분을 제외하는
것을 1순위로 처리하고 +-1 범위 내에서 여분이 있는지 확인 후 여분 list에서 삭제하는 식으로 진행한다.
