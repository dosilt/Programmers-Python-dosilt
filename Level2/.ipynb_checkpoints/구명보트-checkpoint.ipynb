```python
def solution(people, limit):
    p = sorted(people)
    front = 0
    back = len(p)-1
    answer = 0
    while front <= back:
        temp = p[back]
        back -= 1
        if temp + p[front] <= limit:
            front += 1
        answer += 1
    return answer
```

```python
print(solution([70, 50, 80, 50], 100)) # 3
print(solution([70, 80, 50], 100)) # 3
```

# 해설
사람들의 몸무게를 sort한 후 가장 가벼운 사람과 가장 무거운 사람 순으로 한명씩 보트에 태울 때 보트 허용 무게보다

낮으면 더 태우고 아니면 무거운 사람 먼저 옮기는 식으로 옮긴다.
