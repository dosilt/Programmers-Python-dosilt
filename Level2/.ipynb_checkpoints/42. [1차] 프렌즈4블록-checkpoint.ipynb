```python
def find_2_by_2(temp):
    box = set()
    for i in range(1, len(temp)):
        for j in range(1, len(temp[0])):
            if temp[i][j] == temp[i-1][j] == temp[i][j-1] == temp[i-1][j-1] and temp[i][j] != '0':
                box.add((i, j))
                box.add((i-1, j))
                box.add((i, j-1))
                box.add((i-1, j-1))

    for a, b in box:
        temp[a][b] = '0'
        
    return len(box), temp
        
def sorting(temp):
    for i in range(len(temp)):
        for j in range(len(temp[0])):
            if temp[i][j] == '0':
                temp[i].insert(0, temp[i].pop(j))
    return temp
    
def solution(m, n, board): # m 은 행의 개수 n 은 열의 개수
    board = list(map(list, board))
    temp = [['' for i in range(m)] for i in range(n)]
    for i in range(len(board)): # 열
        for j in range(len(board[0])): # 행
            temp[j][i] = board[i][j]
    while True:
        a, temp = find_2_by_2(temp)
        if a != 0:
            temp = sorting(temp)
        else:
            break
    return sum(temp, []).count('0')
```

```python
print(solution(4, 5, ['CCBDE', 'AAADE', 'AAABF', 'CCBBF'])) # 14
print(solution(6, 6, ['TTTANT', 'RRFACC', 'RRRFCC', 'TRRRAA', 'TTMMMF', 'TMMTTJ'])) # 15
```

# 해설
배열에서 열을 기준으로 행 연산을 하는 것보단 ( 세로 연산을 하는 것보단 )

행을 기준으로 열 연산을 하는 것이 ( 가로 연산으로 하는 것이 ) 간단하므로

우선적으로 transpose를 한 후 연산을 진행 하도록 하였다.
