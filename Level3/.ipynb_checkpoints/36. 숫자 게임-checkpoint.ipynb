```python
def solution(A, B):
    A = sorted(A, reverse=True)
    B = sorted(B, reverse=True)
    i, j = 0, 0
    answer = 0
    while i < len(A):
        if A[i] < B[i]:
            answer += 1
        else:
            B.insert(i, B.pop(-1))
        i += 1
    return answer
```

```python
print(solution([5,1,3,7], [2,2,6,8])) # 3
print(solution([2,2,2,2], [1,1,1,1])) # 0
print(solution([1,2,3,4], [1,2,3,4])) # 3
```

# 해설
A의 순서를 알기 때문에 B의 순서만 조절해서 최대로 얻을 수 있는 점수를 계산 하면 된다.

따라서 A와 B를 내림차순으로 정렬 한 후에 B > A 일 경우에는 점수를 +1 하고 B <= A 일 경우에는

필요없는 즉 제일 작은 B의 값을 땡겨와서 비교하고 나머지를 한 칸씩 뒤로 미뤄 비교한다.
