```python
def solution(s):
    answer = 0
    s = list(s)
    for i in range(len(s)):
        temp = s[i]
        for j in range(i, len(s)):
            if s[j] == temp:
                z = 1
                while i+z < j-z:
                    if s[i+z] != s[j-z]:
                        break
                    z += 1

                if i+z >= j-z:
                    if answer < j-i:
                        answer = j-i+1
    return answer
```

```python
print(solution("abcdcba")) # 7
print(solution("abacde")) # 3
```

# 해설
단어가 주어지면 앞에서부터 하나씩 비교를 할 것이다.

예를 들면 1번 예제에 보면 "abcdcba" 라는 문자열이 주어지는데 맨 앞 "a" 를 우선 temp 에 넣고

나머지 문자열 "bcdcba" 에서 앞에서부터 temp랑 같은게 있는지 찾아본다.

a가 맨 뒤에 있는것을 확인후에 z라는 변수를 이용해 앞에서는 하나씩 뒤로 뒤에서는 하나씩 앞으로 다음 문자들이

같은지를 while문 안에서 비교를 한 후 제일 긴 길이만 answer에 저장해 준다.
