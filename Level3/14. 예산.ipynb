```python
def solution(bud, M):
    right = max(bud)
    left = 0
    
    while left <= right:
        temp = 0
        mid = (right + left) // 2
        
        for i in bud:
            if i < mid:
                temp += i
            else:
                temp += mid
        
        if temp < M:
            left = mid+1
        else:
            right = mid-1
            
    return right
```

```python
print(solution([120, 110, 140, 150], 485)) # 127
```
