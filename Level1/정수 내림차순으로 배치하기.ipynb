```python
def solution(n):
    return int(''.join(sorted(list(str(n)), reverse=True)))
```

```python
print(solution(118372)) # 873211
```

# 해설
n 을 문자열로 바꾸어 list형으로 하나씩 쪼개고 순서를 reverse를 해서 내림차순으로 sorted후에 join으로 하나의 문자열로 만들고
int() 를 사용하여 정수로 바꾸어 준다,
