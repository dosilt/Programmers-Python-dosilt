```python
def solution(n):
    if n == 1:
        return 1
    temp = list(range(n+1))
    answer = 0
    for i in range(1, n+1):
        if n % i == 0:
            answer += temp[i]
    return answer
```

```python
print(solution(12)) # 28
print(solution(5)) # 6
```

# 해설
temp라는 n까지의 list를 선언하고 for문을 돌려 나누어 떨어지는 값만 answer에 더해준다.
