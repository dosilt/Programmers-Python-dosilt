```python
from collections import defaultdict

def solution(genres, plays):
    g = defaultdict(list)
    p = defaultdict(int)
    
    for i, data in enumerate(zip(genres, plays)):
        a, b = data
        g[a].append([i, b])
        p[a] += b
    seq = sorted(p, key=lambda x:p[x], reverse=True)
    answer = []
    for i in seq:
        if len(g[i]) >= 2:
            temp = sorted(g[i], key=lambda x: x[1], reverse=True)
        else:
            temp = g[i]
        a = 0
        while a < 2 and a < len(temp):
            answer.append(temp[a][0])
            a += 1
    return answer
```

```python
print(solution(["classic", "pop", "classic", "classic", "pop"], [500, 600, 150, 800, 2500])) # [4, 1, 3, 0]
print(solution(['a', 'b', 'c', 'd', 'e', 'f'], [1, 2, 3, 4, 5, 6]))
print(solution(['classic', 'pop', 'classic', 'classic', 'pop', 'zazz', 'zazz'], [500, 600, 150, 800, 2500, 2100, 1000]))
# [4, 1, 5, 6, 3, 0]
```

# 해설
위 풀이법은 각각의 장르마다 트랙번호와 play수를 넣은 g 라는 dictionary 와

장르마다 총 play 수를 넣은 p 라는 dictionary 2개를 이용하여 총 play 수 변수를 이용하여 제일 많이 실행된

장르 순으로 seq라는 변수를 따로 만들었고 그 순서대로 g 변수에서 play 수에 맞게 큰거부터 2개를 뽑아온다

만약 play 가 1군데 밖에 없었을 경우에는 2개를 뽑아 올수 없기 때문에 밑에 while문을 이용하여

그 경우에는 1개만 뽑아 올수 있도록 하였다.
