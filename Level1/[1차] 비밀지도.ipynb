```python
def solution(n, arr1, arr2):
    answer = [['' for i in range(n)] for i in range(n)]
    tmp1, tmp2 = [], []
    for i, j in zip(arr1, arr2):
        data1 = bin(i)[2:]
        data2 = bin(j)[2:]
        while len(data1) != n:
            data1 = '0' + data1
        while len(data2) != n:
            data2 = '0' + data2
        tmp1.append(data1)
        tmp2.append(data2)
    
    for i in range(n):
        for j in range(n):
            answer[i][j] = int(tmp1[i][j]) + int(tmp2[i][j])
    t = ['#' if b != 0 else ' ' for a in answer for b in a]
    final = []
    final = [''.join(t[i*n:(i+1)*n]) for i in range(n)]
    return final
```

```python
print(solution(5, [9, 20, 28, 18, 11], [30, 1, 21, 17, 28])) # ["#####","# # #", "### #", "# ##", "#####"]
print(solution(6, [46, 33, 33 ,22, 31, 50], [27 ,56, 19, 14, 14, 10])) # ["######", "### #", "## ##", " #### ", " #####", "### # "]
```
