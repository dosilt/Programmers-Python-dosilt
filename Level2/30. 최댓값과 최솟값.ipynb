```python
def solution(s):
    l = list(map(int, s.split(' ')))
    return str(min(l)) + ' ' + str(max(l))
```

```python
print(solution('1 2 3 4')) # '1 4'
print(solution('-1 -2 -3 -4')) # '-4 -1'
print(solution('-1 -1')) # '-1 -1'
```

# 해설
문자열을 공백을 기준으로 list형식으로 변경하고 내부 숫자를 int 형으로 변환한다.

그 이후 최소 값과 최대 값을 추출하여 답의 형식에 맞게 return 해 준다.
