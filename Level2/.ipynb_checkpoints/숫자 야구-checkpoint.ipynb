```python
from collections import Counter

def check(num, rule):
    strike = 0
    ball = 0
    rules = list(str(rule[0]))
    for i in range(3):
        if num[i] == rules[i]:
            strike += 1
        elif num[i] in rules:
            ball += 1
    if strike == rule[1] and ball == rule[2]:
        return True
    return False

def solution(baseball):
    answer = []
    for i in range(123, 987):
        num = list(str(i))
        a = list(Counter(num).keys())
        if len(a) != 3 or '0' in a:
            pass
        else:
            if check(num, baseball[0]):
                answer.append(num)
    for i in baseball[1:]:
        answer = [a for a in answer if check(a, i)]
    return len(answer)
```

```python
print(solution([[123, 1, 1], [356, 1, 0], [327, 2, 0], [489, 0, 1]])) # 2
```

# 해설
일단 우리가 만들 수 있는 숫자를 모두 만들어 놓는다. (최소 123부터 최대 987까지 숫자간의 중복은 없어야 하며 0은 존재할 수 없다.)

그 이후 만들 수 있는 숫자가 너무 많기 때문에 주어진 조건을 다 일일이 부합시키는지 계산해도 되지만

너무 많아서 1번 조건으로 어느 정도 간추리고 하는것도 나쁘지 않다고 생각되어 우선 첫번째 조건을 맞추고

가능한 배열만 따로 뽑아내여 나머지 조건들을 연산하였다.
