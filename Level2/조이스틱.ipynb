```python
def change_alpha(a):
    ccw = ord('Z') - ord(a) + 1
    cw = ord(a) - ord('A')
    return cw if cw < ccw else ccw

def move(name, cursor):
    for i in range(1, len(name)):
        temp_r = cursor+i if cursor+i < len(name) else cursor+i-len(name)
        temp_l = cursor-i if cursor-i >= 0 else cursor-i+len(name)
        if name[temp_r] != 'A':
            return i, temp_r
        elif name[temp_l] != 'A':
            return i, temp_l
    return -1, -1
    
def solution(name):
    n = list(name)
    cursor = 0
    answer = change_alpha(n[cursor])
    n[0] = 'A'
    while cursor != -1:
        i, temp = move(n, cursor)
        cursor = temp
        answer += i
        answer += change_alpha(n[cursor])
        n[cursor] = 'A'
    return answer+1
```

```python
print(solution('JEROEN')) # 56
print(solution('JAN')) # 23
```

# 해설
모든 자리의 알파벳은 A로 시작하며 조이스틱을 아래로 하면 Z방향으로 알파벳 역순으로 계산하고 위로 하면 B 방향으로 원래 방향으로

알파벳이 하나씩 변경된다. 또한 좌우로 움직이면 커서의 위치를 변경할 수 있다. 기본적으로 탐욕 알고리즘을 이용하여

A 가 아닌 가장 가까운 곳을 찾는 알고리즘과

원하는 알파벳이 정순이 빠른지 역순이 빠른지 찾는 알고리즘

총 2가지를 잘 섞어서 코드를 짜면 된다.

A의 ascii 값은 65, Z의 ascii 값은 90 이다.
