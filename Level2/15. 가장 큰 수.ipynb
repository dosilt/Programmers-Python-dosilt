```python
def solution(numbers):
    return str(int(''.join(sorted(map(str, numbers), key=lambda x: x*3, reverse=True ))))
```

```python
print(solution([6, 10, 2])) # 6210
print(solution([3, 30, 34, 5, 9])) # 9534330
```

# 해설
문제의 조건이 0이상 1000 이하의 정수가 numbers의 원소들로 올 수 있기 때문에 최대 값은 1000 최소 값은 0이 된다.

그렇게 때문에 이 숫자들은 문자열로 바꾸어 이 문자열을 *3 하여 3개 씩 쌓으면 서로 비교했을 때 문자열의 위치가 서로

변경되지 않는데 말이 좀 어렵지만 직접 써보면 이해 할 수 있을 것이다.

그 다음에 정렬한 수를 합치고 str(int())를 해주어 0이 맨 앞에 오는 것을 방지할 수 있다.
