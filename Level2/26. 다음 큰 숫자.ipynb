```python
def solution(n):
    temp = bin(n).count('1')
    while True:
        n += 1
        if bin(n).count('1') == temp:
            return n
```

```python
print(solution(78)) # 83
print(solution(15)) # 23
```

# 해설
우선 초기에 주어진 입력 값의 1의 개수를 센다.

이후 주어진 입력값에 1씩 더하면서 처음 값과 1의 개수가 같아 질때 까지 연산을 진행하여

1의 개수가 같을 때의 값을 return 한다.
