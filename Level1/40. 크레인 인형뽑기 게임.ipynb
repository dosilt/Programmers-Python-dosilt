```python
def move(board, i):
    temp = 0
    for n in range(len(board[0])):
        if board[i-1][n] != 0:
            temp = board[i-1][n]
            board[i-1][n] = 0
            return board, temp
    return board, temp

def check(temp, basket, answer):
    if temp != 0:
        if len(basket) < 1:
            basket.append(temp)
        else:
            if basket[-1] == temp:
                basket.pop(-1)
                answer += 2
            else:
                basket.append(temp)
    return basket, answer

def solution(board, moves):
    basket = []
    answer = 0
    board = list(map(list, zip(*board)))
    temp = 0
    for i in moves:
        board, temp = move(board, i)
        basket, answer = check(temp, basket, answer)
    return answer
```

```python
print(solution([[0,0,0,0,0],[0,0,1,0,3],[0,2,5,0,1],[4,2,4,4,2],[3,5,1,3,1]], [1,5,3,5,1,2,1,4])) # 4
```
