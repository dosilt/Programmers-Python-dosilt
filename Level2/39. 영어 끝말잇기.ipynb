```python
def solution(n, words):
    stack = [words[0]]
    error = 0
    for i in range(1, len(words)):
        if words[i][0] != stack[-1][-1]: # 문자 앞 뒤가 다르다면
            error = i
            break
        elif words[i] in stack: # 중복 문자라면
            error = i
            break
        else:
            stack.append(words[i])
    a, b = divmod(error, n)
    if a != 0 or b != 0:
        return [b+1, a+1]
    return [0, 0]
```

```python
print(solution(3, ['tank', 'kick', 'know', 'wheel', 'land', 'dream', 'mother', 'robot', 'tank'])) # [3,3]
print(solution(5, ['hello', 'observe', 'effect', 'take', 'either', 'recognize', 'encourage', 'ensure', 
                   'establish', 'hang', 'gather', 'refer', 'reference', 'estimate', 'executive'])) # [0,0]
print(solution(2, ['hello', 'one', 'even', 'never', 'now', 'world', 'draw'])) # [1,3]
```

# 해설
문자를 하나씩 stack에 넣어가면서 2가지 조건에 맞추어 평가를 진행한다.

만약 error가 발생하면 그때의 index number 를 return 하고 주어진 조건에 따라

solution함수에서 return 한 값을 출력한다.
