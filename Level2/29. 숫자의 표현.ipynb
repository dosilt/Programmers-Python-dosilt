```python
def solution(n):
    temp = [1]
    count = 0
    for i in range(2, n//2+3):
        while sum(temp) > n:
            temp.pop(0)
            
        if sum(temp) == n:
            temp.pop(0)
            count += 1
        temp.append(i)
    return count+1
```

```python
print(solution(15)) # 4
```

# 해설
연속된 자연수를 더하여 n 이라는 숫자를 만드는 것이 문제이다.

따라서 자연수를 지속적으로 temp라는 배열에 집어 넣고 sum(temp) 가 == n을 만족할 때 카운트의 개수를 하나씩

증가 시키면서 자기 자신까지 연산을 진행한다. 자기 자신까지 표현하는데에는 비효율적인 연산이 들어갈 수 있으므로

적당히 절반보다 좀더 큰 범위까지만 연산하고 답에 자기 자신을 포함한 1을 더해주면 속도를 향상 시킬 수 있다.
