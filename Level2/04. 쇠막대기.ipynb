```python
def solution(arrangement):
    stack = []
    answer = 0
    arrangement = arrangement.replace('()', '.')
    for i in range(len(arrangement)):
        if arrangement[i] == '(':
            stack.append('(')
        elif arrangement[i] == '.':
            answer += len(stack)
        else:
            answer += 1
            stack.pop(-1)
    return answer
```

```python
print(solution('()(((()())(())()))(())')) # 17
```

# 해설
우선 레이저 절단기의 위치와 판의 위치를 알 필요가 있다 () 의 경우는 레이저 절단기이며 ( ???? ) 일 경우는 판이라고

생각할 수 있다. 따라서 우선 ()을 '.' 이라는 새로운 문자로 변경한다.

이후에 stack을 이용하여 '(' 일 경우는 append 를 하고 ')'일 경우에는 pop을 하며 동시에 판의 마지막 부분 이기 때문에 +1 을 한다.

'.'을 만났을 경우에는 stack의 길이 즉, 판의 개수 만큼 answer에 더해준다
