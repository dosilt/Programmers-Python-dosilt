```python
def solution(n):
    fibo = [0, 1, 1, 2] + [0] * (n-3)
    if n <= 3:
        return fibo[n]
    else:
        for i in range(4, n+1):
            fibo[i] = fibo[i-2] + fibo[i-1]
        return fibo[n] % 1234567
```

```python
print(solution(3)) # 2
print(solution(5)) # 5
```
