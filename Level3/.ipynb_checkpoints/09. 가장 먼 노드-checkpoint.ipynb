```python
from collections import defaultdict

def solution(n, edge):
    edge = sorted(edge, key=lambda x:x[:])
    visited = [False] * (n + 1)
    dic = defaultdict(list)
    for a, b in edge:
        dic[a].append(b)
        dic[b].append(a)
    q = [1]
    answer = 0
    while q:
        answer = len(q)
        temp = []
        for i in q:
            visited[i] = True
            for j in dic[i]:
                if visited[j] == False:
                    visited[j] = True
                    temp.append(j)
        q = temp
    return answer
```

```python
print(solution(6, [[3, 6], [4, 3], [3, 2], [1, 3], [1, 2], [2, 4], [5, 2]])) # 3
```

# 해설
collections 에 있는 defaultdict(list) 를 이용하여 일반적인 dictionary 자료형과 같지만 기본 default 가 list이기 때문에

바로 append 를 사용할 수 있다는 장점이 있다. 이렇게 모든 노드의 인접 관계에 대한 dict를 만들고

그 이후에 dict[1]부터 bfs 를 이용하여 visited한 곳을 제외하고 1층씩 깊게 들어가게 하였다.
