```python
def solution(phone_book):
    book = sorted(phone_book, key=lambda x: len(x))
    for i in range(len(book)):
        for j in range(i+1, len(book)):
            if book[i] == book[j][:len(book[i])]:
                return False
    return True
```

```python
print(solution(['119', '97674223', '1195524421'])) # False
print(solution(['123', '456', '789'])) # True
print(solution(['12', '123', '1235', '567', '88'])) # False
```

# 해설
전화번호 중에 하나라도 접두어인 경우가 있으면 False 아니면 True를 return 한다.

접두어가 되기 위해선 비교하려는 대상보다 비교하려는 것이 길이가 짧아야 하기 때문에 우선적으로

len(x)에 의한 정렬을 하고 2중 for문을 이용해 무조건 자기보다 우측에 있는 자료들과 비교를 진행 하게 하였다.
