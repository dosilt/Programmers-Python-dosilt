```python
def solution(skill, skill_trees):
    skill_seq = list(skill)
    count = 0
    for i in skill_trees:
        seq = 0
        possible = True
        for j in i:
            if j in skill_seq:
                if j == skill_seq[seq]:
                    seq += 1
                elif seq == len(skill_seq):
                    break
                else:
                    possible = False
                    break
        if possible == True:
            count += 1
    return count
```

```python
print(solution("CBD", ["BACDE", "CBADF", "AECB", "BDA"])) # 2
```

# 해설
우선 우리가 배울 스킬의 순서를 list형식으로 하나씩 쪼개어 놓는다.

skill_trees에 있는 원소를 하나씩 for문으로 돌리면서 배울 스킬이 존재하는지 in을 이용하여 확인한다.

skill_tress에 있는 원소중에 배울 스킬이 포함 되어 있으면 seq를 이용하여 순서대로 배우는 위치가 맞는지 확인한다.

만약 skill_trees에 있는 스킬 순서가 배울 스킬 순서와 다르다면 for문을 break하고 possible을 False로 변경한다.

만약 순서가 맞다면 위의 과정을 계속 반복한다.
