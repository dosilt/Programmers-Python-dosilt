```python
def solution(dartResult):
    count, start = 0, 0
    answer = ['', '', '']
    split = ['', '', '']
    for end in range(1, len(dartResult)):
        try:
            int(dartResult[end-1])
        except:
            try:
                int(dartResult[end])
                split[count] = dartResult[start:end]
                start = end
                count += 1
            except:
                pass
        if count == 2:
            split[count] = dartResult[start:]
    for n, i in enumerate(split):
        for data in i:
            if data == 'S':
                answer[n] += '**1'
            elif data == 'D':
                answer[n] += '**2'
            elif data == 'T':
                answer[n] += '**3'
            elif data == '*':
                if n != 0:
                    answer[n-1] += '*2'
                answer[n] += '*2'
            elif data == '#':
                answer[n] += '*(-1)'
            else:
                answer[n] += data
    return sum([eval(a) for a in answer])
```

```python
print(solution('1S2D*3T')) # 37, 11 * 2 + 22 * 2 + 33
print(solution('1D2S#10S')) # 9, 12 + 21 * (-1) + 101
print(solution('1D2S0T')) # 3, 12 + 21 + 03
print(solution('1S*2T*3S')) # 23, 11 * 2 * 2 + 23 * 2 + 31
print(solution('1D#2S*3S')) # 5, 12 * (-1) * 2 + 21 * 2 + 31
print(solution('1T2D3D#')) # -4, 13 + 22 + 32 * (-1)
print(solution('1D2S3T*')) # 59, 12 + 21 * 2 + 33 * 2
```
