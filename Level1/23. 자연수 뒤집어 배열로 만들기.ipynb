```python
def solution(n):
    temp = []
    while n > 0:
        temp.append(n%10)
        n = n // 10
    return temp
```

```python
print(solution(12345)) # [5,4,3,2,1]
```

# 해설
숫자의 10으로 나눴을 때 나머지를 temp에 저장하고 10으로 나눴을 때 몫을 다음 n으로 집어 넣는다.
