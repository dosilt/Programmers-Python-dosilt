```python
def solution(str1, str2):
    temp1 = []
    temp2 = []
    for i in range(len(str1)-1):
        if str1[i].isalpha() and str1[i+1].isalpha():
            temp1.append(''.join(str1[i:i+2].lower()))
    for i in range(len(str2)-1):
        if str2[i].isalpha() and str2[i+1].isalpha():
            temp2.append(''.join(str2[i:i+2].lower()))
    count = 0
    for i in temp1:
        if i in temp2:
            count += 1
            temp2.remove(i)
            
    low = len(temp1) + len(temp2)
    if low != 0:
        return int(count / low * 65536)
    return 65536
```

```python
print(solution('FRANCE', 'french')) # 16384
print(solution('handshake', 'shake hands')) # 65536
print(solution('aa1+aa2', 'AAAA12')) # 43690
print(solution('E=M*C^2', 'e=m*c^2')) # 65536
```

# 해설
우선 문자열을 2개씩 쪼개야 하며 2개씩 쪼갤때 특수문자는 들어갈 수 없다. 따라서 isalpha()를 이용해서 문자인지 아닌지를 판별하여

문자일 경우만 2개씩 쪼개어 temp라는 변수에 저장을 해둔다. 이후에 temp1과 temp2의 교집합 부분을 구하기 위해 for문을 이용해

temp1 의 데이터가 temp2에 있는지 확인하고 있으면 이 데이터를 지우면서 교집합 변수로 지정한 count 의 값을 1씩 증가시킨다.

교집합 부분을 다 지우고 temp1 과 temp2에 남은 변수들의 길이가 합집합의 길이가 되며 교집합은 count값이 된다.

이 둘을 나누고 65536 을 곱한뒤 int형으로 변경해 준 값을 return 해 주면된다,
