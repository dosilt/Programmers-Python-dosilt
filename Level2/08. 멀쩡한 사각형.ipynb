```python
# from math import gcd

def made_gcd(a, b):
    while a != 0:
        a, b = b % a, a
    return b

def solution(w, h):
#     print(w*h - (w+h-gcd(w,h)))
    return w*h - (w + h - made_gcd(w, h))
```

```python
print(solution(8, 12)) # 80
```

# 해설
대각선 방향으로 선이 하나 존재할 때 멀쩡하게 존재하는 정사각형의 개수를 세는 것이 문제이다.

1 by 1 에서는 0개가 존재하고

2 by 3 에서는 2개, 4 by 6 에는 16개, 5 by 2 에는 4개의 정사각형이 존재한다.

여기서는 가로축과 세로축을 나눠서 생각해 볼 수 있다. 

가로 축에 사용할 수 없는 정사각형의 개수는 가로 축의 길이 만큼이고

세로 축에 사용할 수 없는 정사각형의 개수는 세로 축의 길이 만큼이다.

즉 w + h 만큼을 이용할 수 없다는 것인데 이 중에서 겹치는 공간이 존재하기 때문에 그 겹치는 공간인
