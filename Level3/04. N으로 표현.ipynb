```python
def solution(N, number):
    answer = [0, [N]]
    for i in range(2, 9):
        s = set()
        for j in range(1, i//2+1):
            for a in answer[j]:
                for b in answer[i-j]:
                    s.add(a+b)
                    s.add(a-b)
                    s.add(b-a)
                    s.add(a*b)
                    if b != 0:
                        s.add(a//b)
                    if a != 0:
                        s.add(b//a)
                    s.add(10*N+N)
        if number in s:
            return i
        answer.append(s)
    return -1
```

```python
print(solution(5, 12)) # 4
print(solution(2, 11)) # 3
print(solution(5, 26)) # 4
```

# 해설
이 문제는 N이라는 숫자를 1개 사용했을 때 만들수 있는수 2개를 사용했을 때 만들수 있는 수를 answer이라는 list에 저장하고

number가 생성되는지를 확인하는 if문으로 되있다. for문이 많이 있는 이유를 말하기에 앞서 일단

N은 5라고 가정했을 때 N = 1 -> [5]

N = 2 -> [5+5, 5-5, 5//5, 5*5, 55] 를 만들 수 있다.

즉 이것이 N이 늘어날수록 증가하는데 증가하는 방식이 

N = 3 을 만들기 위해선 N=2일때 값 + N=1일때 값 이면서 반대로 N=1일때의 값 + N=2일때의 값이다.

하지만 반복이기 때문에 이러한 값은 제어하고 대신 빼기나 나누기 연산을 동시에 두었다 (x-y) (y-x) (x//y) (y//x)

N = 5일때를 생각해 보면 N=1 + N=4, N+2 + N+3 의 경우의 수를 모두 고려해야한다.
