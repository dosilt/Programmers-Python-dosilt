```python
def solution(routes):
    routes = sorted(routes, key=lambda x: x[1])
    camera_loc = -30000
    answer = 0
    for route in routes:
        if camera_loc < route[0]:
            answer += 1
            camera_loc = route[1]
    return answer
```

```python
print(solution([[-20,15], [-14,-5], [-18,-13], [-5,-3]])) # 2
```

# 해설
고속도로에 차들이 지나다닐 때 모든 차들이 카메라에 찍히도록 하도록 하는 카메라 위치를 조정해야 한다.

따라서 위 방법으로는 차들이 나가는 위치대로 정렬한 후에

1) 차들이 나가는 순서대로 정렬 했기 때문에 먼저 나간 차를 찍을 카메라는 무조건 존재 해야한다.
    따라서 가장 이상적인 제일 먼저 나가는 차의 위치에 카메라를 단다고 가정한다.

2) 먼저 나간 차가 나간 위치보다 일찍 들어온 차들은 1) 조건에 따른 카메라에 무조건 찍힌다,

3) 1) 조건에 부합하지 않은 제일 먼저 나간 차가 나간 위치보다 들어온 위치가 뒤에서 고속도로에 들어온 차는                           
    새로운 카메라가 필요하다 (위 코드의 answer +=1) 부분
    
4) 새로운 카메라는 위 조건에 부합하지 않으면서 들어온 차가 나간 위치에 설치한다.

5) 위 조건들을 반복한다.
