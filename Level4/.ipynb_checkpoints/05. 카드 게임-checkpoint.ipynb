```python
def solution(left, right):
    dp = [[-1 for _ in range(len(right)+1)] for _ in range(len(left)+1)] # 배열을 저장할 dp 배열 선언
    dp[0][0] = 0 # 처음 위치는 0으로 셋팅
    answer = 0
    for i in range(len(left)): # left 숫자의 카드를 처음부터 하나씩 입력으로 넣는다.
        for j in range(len(right)): # right 숫자의 카드를 처음부터 하나씩 입력으로 넣는다.
            
            if dp[i][j] == -1: # dp[i][j] 가 -1 일 경우는 left의 숫자가 right보다 작을 경우이다.
                continue
                
            if left[i] > right[j]: # left가 right 보다 클 경우에는 right카드를 버리고 dp에 저장한다.
                dp[i][j+1] = max(dp[i][j+1], dp[i][j]+right[j])
                answer = max(answer, dp[i][j+1]) # 지속적인 최대값 추출기
                
            if dp[i][j] > dp[i+1][j]: # left카드를 한장 버렸다고 가정했을 시 얻을 수 있는 최대값을 다음 위치에 저장
                dp[i+1][j] = dp[i][j]
                
            if dp[i][j] > dp[i+1][j+1]: # left카드와 right카드를 한장씩 버렸다고 가정했을 시 얻을 수 있는 최대값을 다음 위치에 저장
                dp[i+1][j+1] = dp[i][j]
    
    return answer
```

```python
print(solution([3, 2, 5], [2, 4, 1])) # 7
```
