```python
def solution(weight):
    weight = sorted(weight)
    s = 1
    for i in weight:
        if i > s:
            return s
        else:
            s += i
    return sum(weight)+1
```

```python
print(solution([3, 1, 6, 2, 7, 30, 1])) # 21
```

# 해설
코딩의 난이도 보단 수학적인 지식이 필요한 문제이다. 일반적인 탐욕법으로 s(물건의 무게) 를 1씩 증가시키며

가장 s와 가까운 추 순으로 저울에 올리며 계산을 하였지만 효율성 test에서 시간초과가 떠버린다.

이 방법 같은 경우는 추의 무게를 계속적으로 물건의 무게에 더해가며 푸는 방식이다.

이해하기 위해선 DP랑 같이 생각하면 편할 것 같다.
