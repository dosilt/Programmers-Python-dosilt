```python
def solution(priorities, location):
    num = list(range(len(priorities)))
    temp = float('inf')
    count = 0
    while location != temp:
        max_ = max(priorities)
        while max_ != priorities[0]:
            data1 = priorities.pop(0)
            data2 = num.pop(0)
            priorities.append(data1)
            num.append(data2)
        priorities.pop(0)
        temp = num.pop(0)
        count += 1
    return count
```

```python
print(solution([2, 1, 3, 2], 2)) # 1
print(solution([1, 1, 9, 1, 1, 1], 0)) # 5
```

# 해설
priorities라는 list에는 우선 순위 배열이 들어가 있으며 숫자가 클수록 우선순위가 높은 것이다.

우선순위가 높은 일부터 처리를 진행하며 만약 현재 작업보다 우선순위가 큰 작업이 뒤에 있을 경우

현재 작업을 맨 뒤로 보내야 한다. !!! (important) 일반 적인 sort연산으로는 불가능하다.

num.pop의 값이 location과 같을 때까지 while문을 돌려서 같은 값이 나오면 while문을 나오면서 count값을
