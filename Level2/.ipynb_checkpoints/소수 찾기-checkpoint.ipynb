```python
from itertools import permutations

def prime(n):
    temp = [False, False] + [True]*(n-1)
    for i in range(2, n//2):
        if temp[i] == True:
            for j in range(i*2, n+1, i):
                temp[j] = False
    return temp

def solution(numbers):
    num = []
    for i in range(len(numbers)):
        num.extend(list(map(int, map(''.join, permutations(numbers, i+1)))))
    num = list(set(num))
    maximum = max(num)
    prime_list = prime(maximum)
    answer = 0
    for i in num:
        if prime_list[i] == True:
            answer += 1
    return answer
```

```python
print(solution('17')) # 3
print(solution('011')) # 2
```

# 해설
우선 주어진 number 카드로 만들수 있는 숫자가 무엇인지를 알아보기 위하여 itertools에 있는 permutations를 이용하여 각각의 카드로

만들 수 있는 경우의 수를 모두 만든 후에 list(set(num)) 을 이용하여 중복된 숫자들을 제거해 준다.

그 이후 최대 값을 가지는 수를 이용하여 에라토스테네스의 체 알고리즘을 이용하여 최대값보다 작은 숫자들이 소수인지 아닌지를

판별 하는 코드를 생성하여 소수일 경우에는 prime_list에 True가 반환되고 아닐 경우에는 False가 반환된다.

따라서 조합하여 만들 수 있는 경우의 수들이 소수인지 prime_list에 넣어보고 True일 경우 answer의 값을 1 증가시키는 식으로

소수의 개수를 더하면 된다.
