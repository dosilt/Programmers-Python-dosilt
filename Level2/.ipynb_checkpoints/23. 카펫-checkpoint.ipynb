```python
def solution(brown, red):
    area = brown + red
    for height in range(1, area//2+1):
        if area % height == 0:
            width = area // height
            if (width-2) * (height-2) == red:
                return [width, height]
```

```python
print(solution(10, 2)) # [4, 3]
print(solution(8, 1)) # [3, 3]
print(solution(24, 24)) # [8, 6]
```

# 해설
카펫의 면적에서 가운데 부분은 red 색이고 외각은 brown 색이다 즉 카펫의 크기 width와 height 를 알고 있다면

brown + red 의 값은 width * height 와 같은 것이고

red 의 값은 (width - 2) * (height - 2) 와 같을 것이다. (2를 빼는 이유는 좌우 2개와 위아래 2개의 모서리 brown 카펫이 존재하기 때문)

위 2조건을 수학적으로 풀어서 계산한다면 충분히 답을 도출해 낼 수 있다.
