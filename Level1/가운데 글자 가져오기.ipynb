```python
def solution(s):
    length = len(s)
    if length % 2 == 0:
        return s[length//2-1:length//2+1]
    else:
        return s[length//2]
```

```python
print(solution('abcde')) # c
print(solution('qwer')) # we
```

# 해설
문자열의 길이가 홀수면 중간이 존재하지만 짝수일 경우는 중간이 존재하지 않아서 2개의 문자를 출력해야 된다. 따라서 문자열의 길이가
홀수인지 짝수인지를 판단하고 출력할 문자의 위치를 지정해주면된다.
