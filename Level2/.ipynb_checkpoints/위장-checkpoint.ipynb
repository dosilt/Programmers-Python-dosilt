```python
def solution(clothes):
    cloth = {}
    for a, b in clothes:
        try:
            cloth[b].append(a)
        except:
            cloth[b] = [a]
    answer = 1
    for i in cloth:
        answer *= (len(cloth[i]) + 1)
        
    answer -= 1
        
    return answer
```

```python
print(solution([['yellow_hat', 'headgear'], ['blue_sunglasses', 'eyewear'], ['green_turban', 'headgear']])) # 5
print(solution([['crow_mask', 'face'], ['blue_sunglasses', 'face'], ['smoky_makeup', 'face']])) # 3
```

# 해설
입력값으로 받은 clothes 에는 [0] 번째 변수에 옷의 이름이 나와있고 [1] 번째 변수에 옷의 종류가 나와있다.

스파이는 다 벗을 순 없고 다른 것들은 안입고 하나만 입어도 된다. 옷을 종류 별로 개수를 정리 했을 때

안입는 case 인 +1을 해주어야 하고 마지막에 정답을 return 할 때에는 다 안입는 경우를 제외해야 하므로 -1 을 해야한다.
