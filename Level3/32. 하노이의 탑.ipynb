```python
def hanoi(n, start, to, mid, answer):
    if n == 1:
        answer.append([start, to])
        return 0
    hanoi(n-1, start, mid, to, answer)
    answer.append([start, to])
    hanoi(n-1, mid, to, start, answer)

def solution(n):
    answer =[]
    hanoi(n, 1, 3, 2, answer)
    return answer
```

```python
print(solution(2)) # [[1, 2], [1, 3], [2, 3]]
print(solution(3)) # [[1, 3], [1, 2], [3, 2], [1, 3], [2, 1], [2, 3], [1, 3]]
print(solution(4)) # 
```
