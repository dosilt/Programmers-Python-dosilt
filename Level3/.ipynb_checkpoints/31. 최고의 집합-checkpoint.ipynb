```python
def solution(n, s):
    answer = []
    if n > s:
        return [-1]
    while n > 1:
        a, b = divmod(s, n)
        s = s - a
        n = n - 1
        answer.append(a)
    answer.append(s)
    return answer
```

```python
print(solution(2, 9)) # [4, 5]
print(solution(2, 1)) # [-1]
print(solution(2, 8)) # [4, 4]
```

# 해설
s 를 n 으로 나누고 이후엔 (s - s//n) / (n-1) 이 식으로 진행하며 n 이 1이 되는 순간 종료된다.

이때까지의 몫(s//n) 을 answer에 append시켜놓고 return 하면 된다.
