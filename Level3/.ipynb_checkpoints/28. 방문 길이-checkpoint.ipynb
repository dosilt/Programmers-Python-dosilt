```python
import copy
def solution(dirs):
    pre_loc = [5, 5]
    cur_loc = [5, 5]
    command = dict({'U':(1, 0), 'L':(0, -1), 'R':(0, 1), 'D':(-1, 0)})
    visited = set()
    
    for i in dirs:
        cur_loc[0] += command[i][0]
        cur_loc[1] += command[i][1]
        if 0 <= cur_loc[0] <= 10 and 0<= cur_loc[1] <= 10:
            visited.add((pre_loc[0], pre_loc[1], cur_loc[0], cur_loc[1]))
            visited.add((cur_loc[0], cur_loc[1], pre_loc[0], pre_loc[1]))
            pre_loc = copy.deepcopy(cur_loc)
        else:
            cur_loc[0] -= command[i][0]
            cur_loc[1] -= command[i][1]
    return len(visited)//2
```

```python
print(solution('ULURRDLLU')) # 7
print(solution('LULLLLLLU')) # 7
print(solution('LR')) # 1
```

# 해설
어렵게 생각하지 말고 우선 움직임을 나타내는 배열 하나를 선언하고 (여기서는 dictionary로 선언) 하나씩 조건에 맞게 불러오면서 이동한

좌표를 visited에 저장만 해주면 된다. visited에 좌표를 저장할때 (전에 있던 좌표, 지금 있는 좌표) 이렇게 2가지를 저장해 주었으며

역으로 (지금 있는 좌표, 전에 있던 좌표) 까지 저장을 해 주어야 한다, 왜냐면 test 3번에서 볼 수 있듯이

왕복은 거리 1로 치기 때문이다. 따라서 역으로 좌표까지 저장을 한뒤 visited의 길이 // 2를 하면 총 이동거리를 알 수 있다.

배열 안의 값을 변경하지 않을 때에는 list형 [] 보단 tuple형 () 으로 선언해야 속도를 향상시킬 수 있다.
