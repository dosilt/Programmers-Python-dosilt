```python
def solution(answers):
    number = [[], [], []]
    number[0] = [1, 2, 3, 4, 5] * (len(answers) // 5 + 1) # 1번 수포자가 찍는 방식
    number[1] = [2, 1, 2, 3, 2, 4, 2, 5] * (len(answers) // 5 + 1) # 2번 수포자가 찍는 방식
    number[2] = [3, 3, 1, 1, 2, 2, 4, 4, 5, 5] * (len(answers) // 5 + 1) # 3번 수포자가 찍는 방식
    number[0] = number[0][:len(answers)] # 문제 갯수와 답안지의 길이를 맞춘다
    number[1] = number[1][:len(answers)]
    number[2] = number[2][:len(answers)]
    temp = [0, 0, 0]
    for i in range(3):
        for j in range(len(answers)):
            if number[i][j] == answers[j]:
                temp[i] += 1
    maximum = max(temp)
    answer = []
    for n in range(len(temp)):
        if temp[n] == maximum:
            answer.append(n+1)
    return answer
```

```python
print(solution([1,2,3,4,5])) # [1])
print(solution([1,3,2,4,2])) # [1,2,3])
```

# 해설
우선 수포자들이 정답을 찍는 방식을 이용하여 문제의 갯수와 같게 만든 후에 2중 for문을 이용하여 각각의 위치에 답이 맞았는지 틀렸는지를
비교하여 답이 맞았을 경우에는 temp[i] 의 값을 1씩 증가시켜 어떤 수포자가 몇개를 맞았는지 최종적으로 저장후에
마지막 answer for문을 이용하여 최대로 맞춘 사람의 번호를 저장시킨다.
