```python
def solution(n, a, b):
    count = 0
    while a != b:
        a = (a+1)//2
        b = (b+1)//2
        count += 1
    return count
```

```python
print(solution(8, 4, 7)) # 3
```

# 해설
1번 2번이 싸워서 1번이 되고

3번 4번이 싸워서 2번이 되고

위의 방식대로 다음 대진표가 형성되기 때문에 위에 알고리즘이 성립하려면 1을 더한뒤 //2 를 진행하면 된다.
