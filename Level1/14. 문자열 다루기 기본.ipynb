```python
def solution(s):
    if len(s) == 4 or len(s) == 6:
        try:
            int(s)
            return True
        except:
            return False
    return False
```

```python
print(solution('a234')) # false
print(solution('1234')) # true
```

# 해설
우선 문제에 주어진 대로 문자열의 길이가 4 또는 6이여야 하기 때문에 처음으로 if문을 사용해 주고 문자열이 숫자로만 되어있으면
True 아니면 False를 return 해 주어야 하기 때문에 try ~ except 문을 이용하여 int로 치환 가능하면 True를 return 해주고
int로 치환할수 없으면 False를 return 해준다. 문자열의 길이가 4 또는 6이 아닐경우도 False를 return 해 준다.
