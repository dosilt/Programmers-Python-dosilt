```python
def solution(progresses, speeds):
    answer = []
    while progresses:
        count = 0
        while progresses and progresses[0] >= 100:
            progresses.pop(0)
            speeds.pop(0)
            count += 1
        if progresses and progresses[0] < 100:
            for i in range(len(progresses)):
                progresses[i] += speeds[i]
        if count != 0:
            answer.append(count)
    return answer
```

```python
print(solution([93,30,55], [1,30,5])) # [2,1]
print(solution([40, 93, 30, 55, 60, 65], [60, 1, 30, 5 , 10, 7])) # [1, 2, 3]
```

# 해설
기능 개발 문제에서 progresses에 있는 데이터가 현재 개발 상황이고 speeds에 있는 데이터가 하루에 개발 속도이다

개발 상황이 100이 되면 배포를 시작하며 배열[0]번이 아닌 다른 쪽에서 먼저 개발이 완료되어도 [0]번의 기능이

개발이 되지 않았다면 배포 되지 못하고 나중에 한번에 개발된다. ( 즉, 왼쪽 기능부터 차례대로 배포 되어야 한다, )

따라서 [0] 번째 위치가 중요하기 때문에 코드 상에선 [0]번째 원소가 100이 넘었는지만 확인하고 넘었다면

speeds와 progresses에서 0번째 원소를 pop하고 위의 과정을 반복적으로 진행한다.
