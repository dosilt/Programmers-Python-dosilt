```python
def fibo(n, answer):
    for i in range(2, n):
        answer[i] = answer[i-2] + answer[i-1]
    return answer

def solution(N):
    answer = [1, 1] +[0] * (N-2)
    answer = fibo(N, answer)
    return answer[-1] * 4 + answer[-2] * 2
```

```python
print(solution(5)) # 26
print(solution(6)) # 42
```
