```python
def solution(s, n):
    temp = ''
    for i in s:
        if i == ' ':
            temp += i
        else:
            o = ord(i)
            if 65 <= o <= 90: # 대문자 일 경우
                while o+n > 90:
                    o -= 26
            else: # 소문자 일경우
                while o+n > 122:
                    o -= 26
            temp += chr(o+n) 
    return temp
```

```python
print(solution('AB', 1)) # BC
print(solution('z', 1)) # a
print(solution('a B z', 4)) # e F d
```

# 해설
ascii 코드표에서 n를 더한 값을 다시 문자로 변환하여 출력하는 것이다. 조심해야 할것은 Z -> A로 z -> a로 변환한다는 것이다.
ascii 코드에서 A = 65, Z = 90, a = 97. z = 122 이다. n을 더했을 때 z나 Z의 범위가 벗어나면 -26을 해주어 다시 a와 A 부터 시작할 수
있도록 코드를 수정해 주면 된다.
