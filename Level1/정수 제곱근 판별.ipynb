```python
from math import sqrt
def solution(n):
    num = sqrt(n)
    if num % 1 == 0: # 정수 판별
        return (num+1)**2
    else:
        return -1
```

```python
print(solution(121)) # 144
print(solution(3)) # -1
```

# 해설
math의 sqrt라는 제곱근을 구하는 함수를 이용하여 제곱근을 구한뒤 num % 1로 나누어 떨어진다면 정수고 아니면 실수인데 정수인지
판별을 하여 정수일 경우에는 +1 한 값에 제곱을 하여 return 하고 정수가 아닐 경우에는 -1을 return 한다.
