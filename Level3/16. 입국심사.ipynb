```python
def solution(n, times):
    right = max(times) * n
    left = 0
    while right > left:
        temp = 0
        mid = (right + left) // 2
        for i in times:
            temp += (mid // i)
        if temp >= n:
            right = mid
        else:
            left = mid+1
    return right
```

```python
print(solution(6, [7, 10])) # 28
print(solution(6, [6,10])) # 24
print(solution(6, [8,10])) # 30
print(solution(6, [4,10])) # 20
print(solution(11, [3,4,10])) # 18
```

# 해설
이분탐색을 통해 이분탐색의 left와 right지점은 걸린 시간이며 구간을 나눌때 사용되는 것은

n의 개수이다. 즉 mid의 시간동안 얼만큼의 작업을 할 수 있는지에 따라 구간을 나눈다.

mid 시간동안 n 보다 많은 일을 하면 right 지점이 mid 가 될 것이며

mid 시간동안 n 보다 적은 일을 하면 left 지점이 mid 가 될 것이다.
