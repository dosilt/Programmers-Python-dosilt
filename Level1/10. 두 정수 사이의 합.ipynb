```python
def solution(a, b):
    return sum(range(a, b+1)) if a < b else sum(range(b, a+1))
```

```python
print(solution(3, 5)) # 12
print(solution(3, 3)) # 3
print(solution(5, 3)) # 12
```

# 해설
a와 b사이의 숫자를 더하는 단순 계산이기 때문에 sum이라는 함수를 이용하여 range로 a부터 b+1까지 (b+1은 포함 안함)
list함수를 더하는 연산을 진행해준다. 기본적으로 range는 range(x, y) 일때 x부터 1씩 더해서 y까지 연산을 해주기 때문에
x가 y보다 클 경우는 값이 1개밖에 되지 않아 y가 크게 해주어야 한다.
