```python
def solution(A,B):
    A = sorted(A)
    B = sorted(B, reverse=True)
    answer = 0
    for i in range(len(A)):
        answer += A[i] * B[i]
    return answer
```

```python
print(solution([1, 4, 2], [5, 4, 4])) # 29
print(solution([1,2], [3,4])) # 10
```

# 해설
A배열과 B배열의 값을 곱해서 더한 값이 최소가 되려면 A의 최소값 * B의 최대값 이런 형식으로 중복 없이 연산을

진행하면 최소값을 만들 수 있다. 따라서 하나는 오름차순으로 정렬하고 하나는 내림차순으로 정렬하여

차례대로 곱한뒤 덧셈을 진행하면 두 배열로 조건에 부합하며 만들 수 있는 최소값을 구할 수 있다.
