```python
def solution(arr1, arr2):
    answer = [[0 for i in range(len(arr2[0]))] for i in range(len(arr1))]
    for i in range(len(answer)): # row
        for j in range(len(answer[0])): # column
            for z in range(len(arr2)):
                answer[i][j] += arr1[i][z] * arr2[z][j]
    return answer
```

```python
print(solution([[1, 4], [3, 2], [4, 1]], [[3, 3], [3, 3]])) # [[15, 15], [15, 15], [15, 15]]
print(solution([[2, 3, 2], [4, 2, 4], [3, 1, 4]], [[5, 4, 3], [2, 4, 1], [3, 1, 1]])) # [[22, 22, 11], [36, 28, 18], [29, 20, 14]]
```

# 해설
행렬의 곱을 연산하기 위해선 우선 A * B 의 행렬 곱이라고 가정하면 A의 column의 수와 B의 row의 수가 같아야 한다.

하지만 문제 조건에 곱 연산이 가능한 행렬만 주어진다고 했기 때문에 위의 가능성은 제외한다.

행렬의 곱에선 A의 행과 B의 열이 서로 곱해져야 하므로 A는 행이 고정되어 있어야 되고 B 는 열이 고정 되어야 한다.

따라서 위의 코드와 같이 arr1[i][z]이며 arr2[z][j] i와 j 는 한번씩만 변하고 z는 지속적으로 변한다.
