```python
import string
def solution(msg):
    msg += 'A'
    temp = string.ascii_uppercase
    data = {}
    for i in range(len(temp)):
        data[temp[i]] = i+1
    answer = []
    start, end = 0, 1
    while end < len(msg):
        while msg[start:end] in data.keys() and end < len(msg):
            end += 1
        
        data[msg[start:end]] = len(data)+1

        answer.append(data[msg[start:end-1]])
        start = end-1
        
    return answer
```

```python
print(solution('KAKAO')) # [11, 1, 27, 15]
print(solution('TOBEORNOTTOBEORTOBEORNOT')) # [20, 15, 2, 5, 15, 18, 14, 15, 20, 27, 29, 31, 36, 30, 32, 34]
print(solution('ABABABABABABABAB')) # [1, 2, 27, 29, 28, 31, 30]
```

# 해설
압축방법은 링크를 이용하여 숙지하면 좋을 것 같다.

우선 기본 대문자 배열을 temp 라는 list형에 저장해 두고 data라는 dictionary 자료형을 제작하여 위 사이트의

형태와 같게 데이터를 정렬 하였다. 이후에 msg에 온 데이터를 비교해 가며 dictionary에 없을 때까지 데이터를

읽다가 없으면 data에 추가하는 식으로 반복적으로 진행 하였다. 여기서 사용한 트릭중에 하나는

맨 처음에 msg에 'A'라는 문자를 더해줬다는 것인데 사실 아무 문자나 더해 줘도 상관이 없다. 더해주지 않을 경우

while문의 조건이 end < len(msg) 이고 answer.append(data[msg[start:end-1]]) 이 문에서 문제가 발생한다.

end - 1 이기 때문에 마지막 전 문자까지만 처리를 진행하게 되는 것이다. 이 문제를 해결 하기 위해

아무 문자나 하나 더해줘서 문제를 해결하였다.
