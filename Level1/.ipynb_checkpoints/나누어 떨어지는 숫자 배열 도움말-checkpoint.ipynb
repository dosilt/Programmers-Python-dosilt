```python
def solution(arr, divisor):
    temp = []
    for i in arr:
        if i % divisor == 0:
            temp.append(i)
    if len(temp) == 0:
        return [-1]
    return sorted(temp)
```

```python
print(solution([5, 9, 7, 10], 5)) # [5, 10]
print(solution([2, 36, 1, 3], 1)) # [1, 2, 3, 36]
print(solution([3, 2, 6], 10)) # [-1]
```

# 해설
arr의 원소들의 divisor로 나누어 떨어지는지를 확인해야 하기 때문에 for문을 이용하여 일일이 하나씩 연산을 진행하여
나누어 떨어지는 수만 temp라는 list에 집어 넣어 놓는다. for문이 끝난 후에 출력은 sort가 되어 있어야 하므로
최종 결과 list에 sort를 진행한다.
