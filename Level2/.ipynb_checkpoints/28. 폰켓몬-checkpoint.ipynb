```python
def solution(nums):
    n = len(set(nums))
    l = len(nums) // 2
    if n > l:
        return l
    return n
```

```python
print(solution([3,1,2,3])) # 2
print(solution([3,3,3,2,2,4])) # 3
print(solution([3,3,3,2,2,2])) # 2
```

# 해설
전체 배열 nums의 길이의 절반 만큼의 포켓몬을 가져갈 수 있다고 할 때 몇 종류의 포켓몬을 가져 갈 수 있는지에 대해

묻는 문제이다. 즉 set 집합을 이용하여 중복을 제거하고 n 이라는 변수에 중복을 제거한 배열의 길이를 저장한 후

l 이라는 변수에 총 몇 마리를 가져갈 수 있는지 저장한다.

그 이후 n > l이면 (가져갈 수 있는 포켓몬의 수가 포켓몬의 종류보다 작으면) l 을 return 하고

그 반대의 경우이면 n 을 return 한다.
