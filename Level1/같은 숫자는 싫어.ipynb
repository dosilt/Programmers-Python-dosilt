```python
def solution(arr):
    stack = [arr.pop(0)]
    for i in arr:
        if stack[-1] != i:
            stack.append(i)
    return stack
```

```python
print(solution([1,1,3,3,0,1,1])) # [1,3,0,1]
print(solution([4,4,4,3,3])) # [4,3]
```

# 해설
같은 숫자인지 비교하기 위해 stack 자료형을 이용할 것이다. 우선 제일 앞에 숫자는 무엇이든 관계없이 stack에 쌓여야 하기 때문에
미리 stack에 넣어두고 arr에 있는 다음 숫자가 stack에 가장 위에 있는 숫자와 같은지 비교하여 다르면 추가로 쌓고 다르면
pass 하는 식으로 진행한다.
