```python
def solution(distance, rocks, n):
    answer = []
    rocks = sorted(rocks)
    for i in range(len(rocks)-1):
        for j in range(i+1, len(rocks)):
            temp1 = rocks[i]
            temp2 = rocks[j]
            
            rocks.remove(rocks[j])
            rocks.remove(rocks[i])
            
            temp = [0] + rocks + [distance]
            ans = float('inf')
            for a in range(1, len(temp)-1):
                ans = min(ans, temp[a] - temp[a-1])
            answer.append(ans)
            
            rocks.insert(i, temp1)
            rocks.insert(j, temp2)
            
    return max(answer)
```

```python
print(solution(25, [2, 14, 11, 21, 17], 2)) # 4
```
