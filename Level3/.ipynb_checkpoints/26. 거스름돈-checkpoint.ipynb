```python
def solution(n, money):
    b = [[0 for _ in range(n+1)] for _ in range(len(money))]
    for i in range(money[0], n+1, money[0]):
        b[0][i] = 1
    for i in range(1, len(money)):
        for j in range(n+1):
            if j == 0:
                b[i][j] = 1
            elif money[i] <= j:
                b[i][j] = b[i-1][j] + b[i][j-money[i]]
            else:
                b[i][j] = b[i-1][j]
    return b[-1][-1] % 1000000007
```

```python
print(solution(5, [1, 2, 5])) # 4
```

# 해설
전형적인 dp를 이용한 문제이다. 위 코드가 이해가 안간다면 간단한 표를 그려서 위의 코드의 순서를 따라

하나씩 그려보면 이해가 될 것이다.
