```python
from collections import deque
def solution(bridge_length, weight, truck_weights):
    b = deque([0 for i in range(bridge_length)])
    b[0] = truck_weights.pop(0)
    count = 1
    while truck_weights:
        b.pop()
        if len(truck_weights) != 0:
            if sum(b) + truck_weights[0] <= weight:
                b.insert(0, truck_weights.pop(0))
            else:
                b.insert(0, 0)
        count += 1
    return count + bridge_length
```

```python
print(solution(2, 10, [7,4,5,6])) # 8
print(solution(100, 100, [10])) # 101
print(solution(100, 100, [10,10,10,10,10,10,10,10,10,10])) # 110
```

# 해설
우선 다리위에 올라올 수 있는 중량 제한이 있기 때문에 다리 길이 만큼의 queue list를 만들어서 다리 위에 중량이 얼만큼 있는지 확인
할 수 있게 하였고, 중량이 초과하지 않는다면 추가적으로 트럭이 올라올 수 있게 하였다. 속도를 향상시키기 위해 일반적인 list 자료형이
아닌 colletions에 있는 deque를 이용하였으며 마지막 트럭이 다리에 올라오면 다리 길이 만큼을 더하여 부가적인 연산을 삭제했다.
