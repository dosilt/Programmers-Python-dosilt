```python
def solution(words, queries):
    answer = [0 for _ in range(len(queries))]
    for i, query in enumerate(queries):
        ans = 0
        length = len(query)
        if query[0] == '?':
            temp = query.split('?')[-1]
            for word in words:
                if length == len(word) and word.endswith(temp):
                    ans += 1
        else:
            temp = query.split('?')[0]
            for word in words:
                if length == len(word) and word.startswith(temp):
                    ans += 1
        answer[i] = ans
    return answer
```

```python
print(solution(["frodo", "front", "frost", "frozen", "frame", "kakao"], ["fro??", "????o", "fr???", "fro???", "pro?"]))
# [3, 2, 4, 1, 0]
```
