```python
import heapq
def solution(n, works):
    for i in range(len(works)):
        works[i] *= -1
    heapq.heapify(works)
    for i in range(n):
        x = heapq.heappop(works)
        if x >= 0:
            return 0
        heapq.heappush(works, x+1)
    answer = 0
    for i in works:
        answer += i*i
    return answer
```

```python
print(solution(4, [4, 3, 3])) # 12
print(solution(1, [2, 1, 2])) # 6
print(solution(3, [1, 1])) # 0
```

# 해설
일일이 배열을 비교해서 최소값을 꺼내기에는 배열이 길어지면 시간이 너무 오래 걸린다. 따라서 heapq 방식의 자료형을 선언하여

works에서 가장 큰 값을 하나씩 불러와서 -1 하는 방식으로 해결해 주었으며, 모든 값이 0보다 크거나 같으면 종료한다.
