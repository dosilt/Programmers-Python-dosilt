```python
from collections import defaultdict
import re
def score(link_d, meta_d, temp_d):
    ans = []
    for i in link_d:
        temp = link_d[i]
        if len(meta_d[i]) != 0:
            for j in meta_d[i]:
                if len(temp_d[j]) != 0:
                    temp += (link_d[j] / len(temp_d[j]))
        ans.append(temp)
    return ans
    
def solution(word, pages):
    link_dict = defaultdict(int)
    meta_dict = defaultdict(list)
    temp_dict = defaultdict(list)
    for i in range(len(pages)):
        link = re.search('"https:.*"', pages[i][re.search('<head>', pages[i]).end():re.search('</head>', pages[i]).start()]).group()
        data = re.split('[^a-z]', pages[i][re.search('<body>', pages[i]).end():re.search('</body>', pages[i]).end()].lower())
        data = data.count(word.lower())
        meta = re.findall('"https:.*"', pages[i][re.search('<body>', pages[i]).end():re.search('</body>', pages[i]).end()])
        link_dict[link] = data
        for i in meta:
            meta_dict[i].append(link)
        temp_dict[link] = meta
    answer = score(link_dict, meta_dict, temp_dict)
    return answer.index(max(answer))
```
