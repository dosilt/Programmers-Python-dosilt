```python
def solution(record):
    ID = {}
    answer = []
    for i in record:
        data = i.split(' ')
        if data[0] == 'Enter': # 들어온 경우
            ID[data[1]] = data[2]
            answer.append(['님이 들어왔습니다.', data[1]])
        elif data[0] == 'Leave': # 나간 경우
            answer.append(['님이 나갔습니다.', data[1]])
        
        else: # 아이디를 변경한 경우
            ID[data[1]] = data[2]
    for i in range(len(answer)):
        answer[i] = ID[answer[i][1]]+answer[i][0]
    return answer
```

```python
print(solution(["Enter uid1234 Muzi", "Enter uid4567 Prodo","Leave uid1234","Enter uid1234 Prodo","Change uid4567 Ryan"]))
# ["Prodo님이 들어왔습니다.", "Ryan님이 들어왔습니다.", "Prodo님이 나갔습니다.", "Prodo님이 들어왔습니다."]
```

# 해설
위 문제의 중요한 점은 아이디는 변경이 불가능하고 이름은 언제나 변경할 수 있다는 것이다.

따라서 해시 자료형을 이용해 아이디명: 이름 순으로 저장을 하여 같은 아이디에 이름이 변경될 때마다

해시 자료형 안의 변수를 변경하여 아이디가 가진 이름이 통일될 수 있도록 해시 자료형에 유지를 해놓고

마지막에 출력할 때 최종적으로 아이디가 지닌 이름으로 한번에 출력이 될 수 있도록 한다.
