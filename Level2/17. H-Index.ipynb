```python
def solution(citations):
    c = sorted(citations)
    l = len(c)
    for i in range(l):
        if c[i] >= l-i:
            return l-i
    return 0
```

```python
print(solution([3, 0, 6, 1, 5])) # 3
print(solution([0, 1, 2, 1]))
```
