```python
def solution(arr):
    if len(arr) == 1:
        return [-1]
    mini = min(arr)
    arr.remove(mini)
    return arr
```

```python
print(solution([4,3,2,1])) # [4,3,2]
print(solution([10])) # [-1]
```

# 해설
list가 빌 경우는 [-1]를 return 하는데 이 경우는 arr의 len이 1일 경우이다. 따라서 위의 경우를 미리 if문을 이용하여 제거하고
arr의 최소값을 mini에 저장후 list형에 이용되는 remove라는 함수를 이용하여 mini와 값이 같은 것을 삭제한다.
