```python
def solution(s):
    stack = [s[0]]
    for i in s[1:]:
        if stack and i == stack[-1]:
            stack.pop(-1)
        else:
            stack.append(i)
    if len(stack) == 0:
        return 1
    else:
        return 0
```

```python
print(solution('baabaa')) # 1
print(solution('cdcd')) # 0
```

# 해설
stack이라는 자료형을 이용하여 뒷 문자와 같으면 stack와 문자를 제거하고 아닐경우엔

stack에 다음 문자를 지속적으로 쌓는 식으로 진행하여 stack이 비었는지 남아있는지로 모두 지워졌는지

아니면 남았는지를 확인할 수 있다.
