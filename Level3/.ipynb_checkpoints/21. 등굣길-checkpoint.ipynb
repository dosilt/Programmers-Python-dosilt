```python
def solution(col, row, puddles):
    road = [[0 for i in range(col+1)] for i in range(row+1)]
    for i in range(1, row+1):
        for j in range(1, col+1):
            if i == 1 and j == 1:
                road[i][j] = 1
                continue
            if [j, i] in puddles:
                road[i][j] == 0
            else:
                road[i][j] = road[i-1][j] + road[i][j-1]
                
    return road[-1][-1] % 1000000007
```

```python
print(solution(4, 3, [[2, 2]]))
```

# 해설
문제를 잘 읽는 사람이라면 상관 없겠지만 그렇지 않다면 문제가 주어진 조건을 잘 읽어봐야 되는 문제이다.

우선 매개변수로 주어지는 m과 n 은 col과 row 순서로 주어지고 puddles안에 있는 좌표도 [y, x] 순으로 주어지기 때문에

코드내에서 row, col을 transpose 하고 계산 하는 사람이 아니고는 좌표 순서를 조심히 해야될 것이다.

마지막에 나눠주는 것도 까먹어서는 안된다.

문제는 단순히 dp 식으로 연산하였으며 puddles 위치일 경우에는 0으로 변경하였다.
