```python
import heapq

def solution(n, cores):
    h = []
    for i in range(len(cores)):
        heapq.heappush(h, (0, i))
    a = 0
    for i in range(n):
        a = heapq.heappop(h)
        heapq.heappush(h, (a[0]+cores[a[1]], a[1]))
    return a[1]+1
```

```python
print(solution(6, [1,2,3])) # 2
print(solution(10, [2, 3]))
```
