```python
def bfs(str1, str2, visited):
    temp = []
    for word1 in str1:
        for num, word2 in enumerate(str2):
            if visited[num] == False:
                test = 0
                for i in range(len(word1)):
                    if word1[i] != word2[i]:
                        test += 1
                if test == 1:
                    temp.append(word2)
                    visited[num] = True
    return temp, visited

def solution(begin, target, words):
    if target not in words:
        return 0
    visited = [False] * len(words)
    temp = [begin]
    answer = 0
    while True:
        temp, visited = bfs(temp, words, visited)
        answer += 1
        if target in temp:
            return answer
```

```python
print(solution('hit', 'cog', ['hot', 'dot', 'dog', 'lot', 'log', 'cog'])) # 4
print(solution('hit', 'cog', ['hot', 'dot', 'dog', 'lot', 'log'])) # 0
```

# 해설
begin 단어로부터 시작해 만들 수 있는 모든 단어를 temp라는 list에 넣어서 bfs 방식으로 한 층씩 쌓아가는 방법이다.
