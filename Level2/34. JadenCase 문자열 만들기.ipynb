```python
def solution(s):
    return ' '.join([a.capitalize() for a in s.split(' ')])
```

```python
print(solution('3people unFollowed me')) # 3people Unfollowed Me
print(solution('for the last week')) # For The Last Week
```

# 해설
파이썬에는 첫문자만 대문자로 변경해주는 capitalize() 라는 내장 함수가 존재한다.

따라서 우선 split으로 나눈후에 capitalize() 를 이용해 첫번째 문자만 대문자로 변경후

다시 join을 이용하여 묶으면 된다.
