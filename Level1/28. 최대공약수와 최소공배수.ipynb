```python
def gcd(a, b):
    while a != 0:
        tmp = a
        a = b % a
        b = tmp
    return b

def solution(n, m):
    tmp = gcd(n, m)
    return [tmp, n*m//tmp]
```

```python
print(solution(3, 12)) # [3, 12]
print(solution(2, 5)) # [1, 10]
```

# 해설
최소 공약수와 최대 공배수를 구해야 하는 문제이며 from math import gcd 로 최대 공약수를 구할 수도 있으나
직접 코드를 구현해 봤으며 최대 공배수는 두 수를 곱하고 최대 공약수로 나누어 주면 된다.
