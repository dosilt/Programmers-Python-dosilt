```python
def solution(n):
    s = str(n)
    answer = 0
    for i in s:
        answer += int(i)
    return answer
```

```python
print(solution(123)) # 6
print(solution(987)) # 24
```

# 해설
각각의 자리수를 더하기 위해 //10 을 이용하는 방법도 있지만 직관적으로 보기엔 string형으로 변경하여 각각의 문자를 더하는 방법이
보기에도 쉽고 이해하기도 쉽기 때문에 str로 바꾼후에 for문으로 하나씩 더하는 방법을 이용했다.
