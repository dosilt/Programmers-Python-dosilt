```python
import heapq
from collections import deque
def solution(stock, dates, supplies, k):
    h = []
    answer = 0
    dates, supplies = deque(dates), deque(supplies)
    while stock < k:
        while dates and dates[0] <= stock:
            date = dates.popleft()
            supply = supplies.popleft()
            heapq.heappush(h, (-supply, supply))
        stock += heapq.heappop(h)[1]
        answer += 1
    return answer
```

```python
print(solution(4, [4, 10, 15], [20, 5, 10], 30)) # 2
```

# 해설
라면 공장에서는 하루에 밀가루를 1톤씩 사용합니다. 원래 밀가루를 공급받던 공장의 고장으로 앞으로 k일 이후에야 밀가루를 공급받을 수 있

기 때문에 해외 공장에서 밀가루를 수입해야 합니다.

라는 것이 문제인데 이 문제에서 중요한 포인트는 밀가루가 떨어지지 않으면서 한번 받을 때 최대한 많이 받는 것이 포인트이다.

따라서 stock으로 버틸 수 있을 만큼의 date 기간 내에서 한번에 받을 수 있는 경우를 heap 자료형을 통해 연산해내어

stock의 총 합이 k보다 커지는 순간 공급을 그만 받아도 된다.
