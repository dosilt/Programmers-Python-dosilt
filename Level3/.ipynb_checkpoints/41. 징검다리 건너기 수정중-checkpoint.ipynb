```python
def solution(stones, k):
    i = 0
    temp = float('inf')
    while i+k <= len(stones):
        data = stones[i:i+k]
        m = max(data)
        index = data.index(m)
        temp = min(m, temp)
        i += index+1
    return temp
```

```python
print(solution([2, 4, 5, 3, 2, 1, 4, 2, 5, 1], 3)) # 3
```
