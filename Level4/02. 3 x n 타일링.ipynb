```python
def solution(n):
    m = n//2
    dp = [1, 3]
    if m < 2:
        return dp[m]
    dp += [0] * (m-1)
    for i in range(2, m+1):
        dp[i] = (3*dp[i-1] + sum(dp[:i-1]*2)) % 1000000007
    return dp[m]
```

```python
print(solution(4)) # 11
print(solution(6))
```
