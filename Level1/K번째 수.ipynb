```python
def solution(array, commands):
    answer = []
    for start, end, index in commands:
        temp = sorted(array[start-1:end])
        answer.append(temp[index-1])
    return answer
```

```python
print(solution([1, 5, 2, 6, 3, 7, 4], [[2, 5, 3], [4, 4, 1], [1, 7, 3]])) # [5, 6, 3]
```

# 해설
commands 안에 있는 값들을 start end index 라는 각각의 변수로 받은 뒤에 array를 start와 end 에 따라 자르고
sorted를 이용해 정렬한 후에 찾고자 하는 index 위치에 있는 값을 answer list에 하나씩 append한다.
