```python
def solution(heights):
    stack = [[heights[-1], len(heights)-1]]
    answer = [0 for i in heights]
    top = 1
    for i in range(len(heights)-2, -1, -1):
        while stack and heights[i] > stack[-1][0]:
            answer[stack[-1][1]] = i+1
            end = i+1
            stack.pop(-1)
        stack.append([heights[i], i])
    return answer
```

```python
print(solution([6,9,5,7,4])) # [0,0,2,2,4]
print(solution([3,9,9,3,5,7,2])) # [0,0,0,3,3,3,6]
print(solution([1,5,3,6,7,6,5])) # [0,0,2,0,0,5,6]
```

# 해설
stack이라는 자료 구조를 이용 할 수 있는가에 대한 문제이다. 문제에서는 오른쪽의 송전탑이 수신하는 데이터는

오른쪽의 송전탑보다 높은 곳에 있는 좌측 송전탑에서만 수신이 가능하다. 즉, 우리는 list를 거꾸로 생각하여

더 높은 송전탑의 index number를 계산해 낼 수 있으면 된다.

stack이라는 변수에 맨 우측에 있는 송전탑의 높이와 index를 저장해 놓는다. 그 이후 좌측으로 하나씩 비교를 한다.

좌측의 송전탑의 높이가 우측(현재 stack에 있는 송전탑)의 높이 보다 작을 경우에는 위의 자료형 처럼 [높이, index] 순으로

stack에 append를 이용해 추가해 놓는다

좌측의 송전탑의 높이가 우측의 높이 보다 높을 경우는 stack에 젤 위에 있는 자료를 빼내어 빼낸 자료의 index number에 대응하는

answer[index] 에 현재 송전탑의 index number를 저장하고 stack에 2개 이상의 송전탑이 존재 할 수도 있으므로 while문을 이용해

지속적으로 비교를 진행한다.
