```python
def solution(n):
    temp = [1, 2] + [0]*(n-2)
    for i in range(2, n):
        temp[i] = temp[i-2] + temp[i-1]
    return temp[-1]%1234567
```

```python
print(solution(4)) # 5
print(solution(3)) # 3
```

# 해설
피보나치 수열과 같은 방식의 풀이 방법을 이용한다.
