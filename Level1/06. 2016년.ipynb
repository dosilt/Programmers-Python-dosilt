```python
def solution(a, b):
    months = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    days = ['FRI','SAT', 'SUN', 'MON', 'TUE', 'WED', 'THU']
    total = sum(months[:a-1], b)
    x, y = divmod(total, 7)
    return days[y-1]
```

```python
solution(5, 24) # TUE
```

# 해설
2016년 1월 1일은 금요일이다. 그러므로 시작 요일은 금요일이 된다. 우선적으로 정리하기에 월마다 몇 일까지 존재하는지 list를 만들고
요일도 금요일부터 순서대로 만든후에 a월 b일까지 총 몇 일인지 일수를 더하고 요일을 구한다
