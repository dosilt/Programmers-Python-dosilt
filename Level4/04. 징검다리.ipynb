```python
def solution(distance, rocks, n):
    rocks.sort()
    left = 0
    right = distance
    answer = 0
    while left <= right:
        mid = (left + right) // 2
        mins = float('inf')
        remove = 0
        prev = 0
        for i in range(len(rocks)):
            if rocks[i] - prev < mid:
                remove += 1
            else:
                mins = min(mins, rocks[i]-prev)
                prev = rocks[i]
        if remove > n:
            right = mid-1
        else:
            answer = mins
            left = mid+1
    return answer
```

```python
print(solution(25, [2, 14, 11, 21, 17], 2)) # 4
```
