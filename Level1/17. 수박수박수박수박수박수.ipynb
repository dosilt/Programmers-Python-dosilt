```python
def solution(n):
    answer = ''
    for i in range(1, n+1):
        if i % 2 == 0:
            answer += '박'
        else:
            answer += '수'
    return answer
```

```python
print(solution(3)) # 수박수
print(solution(4)) # 수박수박
```

# 해설
문자열의 길이가 n이여야 하면서 '수' 와 '박' 이 번갈아 가면서 나오게 출력문을 만들어 주면 된다.
