```python
def gcd(a, b):
    while b != 0:
        a, b = b, a%b
    return a

def solution(arr):
    for i in range(len(arr)-1):
        arr[i+1] = arr[i] * arr[i+1] / gcd(arr[i], arr[i+1])
    return int(arr[-1])
```

```python
print(solution([2,6,8,14])) # 168
print(solution([1,2,3])) # 6
```

# 해설
a, b 의 최소 공배수를 구하기 위해선 a * b / gcd(a, b) 라는 공식을 우선 적으로 이해하고 있어야 풀기 수월한 문제이다.

따라서 위 방법을 이용하여 arr이라는 배열에 있는 숫자들을 2개씩 차례대로 위 공식에 대입하여 최종 출력값을

return 해 주었다.
