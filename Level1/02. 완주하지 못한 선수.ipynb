```python
from collections import Counter
def solution(participant, completion):
    data = Counter(participant) - Counter(completion)
    return list(data.keys())[0]
```

```python
solution(['leo', 'kiki', 'eden'], ['eden', 'kiki']) #leo
solution(['marina', 'josipa', 'nikola', 'vinko', 'filipa'], ['josipa', 'filipa', 'marina', 'nikola']) #vinko
solution(['mislav', 'stanko', 'mislav', 'ana'], ['stanko', 'ana', 'mislav']) #mislav
```

# 해설
파이썬에서 제공하는 함수인 Counter라는 함수를 이용하여 participant와 completion의 list형 자료를 dict형 자료로 바꾸고 각각의 카운트
하여 그 차를 구한다. 만약 참가자가 완주자 명단에도 있을 경우 서로 빼면 사라질 것이고 없으면 남게 된다.
