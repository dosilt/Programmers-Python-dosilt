```python
def solution(n, k):
    answer = [0] * n
    temp = [1] * n
    l = list(range(1, n+1))
    
    for i in range(1, n):
        temp[i] = temp[i-1] * i
    i = 1
    while i < n+1:
        a = (k-1)//temp[-i]
        k = k%temp[-i]
        answer[i-1] = l[a]
        l.remove(l[a])
        i += 1
        
    return answer
```

```python
print(solution(3, 5)) # [3, 1, 2]
```

# 해설
n = 3 즉) [1, 2, 3] 3개의 숫자를 이용하겠다는 의미고 k = 5 즉 [1, 2, 3] 으로 만들 수 있는

숫자중에 사전순으로 5번째 오는 수를 구하라는 문제이다.

따라서 위 코드에서는 answer 는 말그대로 답을 저장하기 위한 배열을 선언하였고

temp는 n! (factorial)을 저장하기 위한 배열로 사용되었다. 위 문제에서 factorial이 필요한 이유는 위의 예제를 예로 들자면

1 [23, 32]
  
2 [13, 31]
  
3 [12, 21]

이런식으로 나열이 가능한데 (k-1) // (n-1)!  의 값이 [1, 2, 3] 배열에서 index 번호가 되며

이후 진행될 k 의 값으로는 k = k % (n-1)! 로 k가 변경된다 따라서 temp에는 최대 (n-1)! 까지 들어가 있다.
