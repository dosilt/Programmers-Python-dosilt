```python
def solution(s):
    answer = 0
    for i in s:
        if i == '(':
            answer += 1
        else:
            answer -= 1
        if answer < 0:
            return False
    if answer == 0:
        return True
    else:
        return False
```

```python
print(solution('()()')) # True
print(solution('(())()')) # True
print(solution(')()()')) # False
print(solution('(()(')) # False
```

# 해설
괄호가 ( 이면 answer 값을 +1 하고 

) 이면 answer 값을 -1 하는 방식으로

answer 이 음수가 될경우 False를 return 하고

answer 이 음수가 아니여도 마지막 연산까지 다 했을 때 answer의 값이 0이 아니면 False를 return

0일 경우에는 True를 return 한다.
