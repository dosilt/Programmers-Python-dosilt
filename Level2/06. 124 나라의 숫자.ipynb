```python
def solution(n):
    n -= 1
    if n == 0:
        return '1'
    num = []
    while n >= 0:
        a, b = divmod(n, 3) # a = 몫, b = 나머지
        n = a - 1
        num.insert(0, str(b))
    answer = ''.join(num)
    answer = answer.replace('2', '4')
    answer = answer.replace('1', '2')
    answer = answer.replace('0', '1')
    return answer
```

```python
print(solution(10)) # 41
print(solution(13)) # 111
print(solution(500000000)) # 421211211121241112
```

# 해설
우리가 사용 할 수 있는 숫자는 1, 2, 4라는 3가지의 숫자밖에 사용 할 수 없다. 즉 3진법의 숫자와 같다고 생각할 수 있지만

3진법은 1의 자리는 0, 1, 2 라는 3가지의 숫자가 이용 가능하지만 두번째 자리 이상부터는 0이라는 숫자가 수학적으로

불가능 하므로 1, 2이라는 2가지 밖에 사용이 안된다. 따라서 이 부분에 있어서 조금의 trick이 필요하다.

0이라는 숫자를 표기하는데 문제가 있기 때문에 우선 초기 10진수 값에 -1 을 하여 연산을 진행하며 나누고 몫이 -1 의 값이

음수가 나올 때까지 진행한다. num이라는 데이터를 쌓는데 있어서 1의 자리 수부터 점차적으로 올라가야하기 때문에

append가 아닌 insert(0, str(b))를 이용하여 무조건 앞에서 부터 데이터를 쌓을 수 있도록 하였다.

마지막에는 이제 우리가 가정했던 방법대로 숫자를 1, 2, 4에 맞게 바꿔주면 된다. 바꾸는 순서에 유의하면서 바꾼다.
