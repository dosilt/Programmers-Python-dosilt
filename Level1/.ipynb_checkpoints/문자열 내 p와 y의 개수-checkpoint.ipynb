# https://programmers.co.kr/learn/courses/30/lessons/12916

```python
def solution(s):
    s = s.lower()
    if s.count('p') == s.count('y'):
        return True
    else:
        return False
```

```python
print(solution('pPoooyY')) # true
print(solution('Pyy')) # false
```

# 해설
대소문자는 구별하지 않고 p의 개수와 y의 개수를 비교하는 것이기 때문에 모든 알파벳을 소문자로 우선만든다.
그 이후에 s라는 문자열에서 p의 개수와 y의 개수가 같으면 True를 return 하고 아닐시엔 False를 return 한다
