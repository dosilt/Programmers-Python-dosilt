```python
from itertools import combinations
def prime(a):
    temp = [0, 0] + [1]*(a-1)
    for i in range(2, a//2+1):
        if temp[i] == 1:
            for j in range(i*2, a+1, i):
                temp[j] = 0
    return temp

def solution(nums):
    n = list(map(sum, (combinations(nums, 3))))
    prim = prime(max(n))
    answer = 0
    for i in n:
        answer += prim[i]
    return answer
```

```python
print(solution([1,2,3,4])) # 1
print(solution([1,2,7,6,4])) # 4
```

# 해설
nums 에 주어진 숫자중에 3개를 더해서 만들 수 있는 수를 우선 combinations 를 이용하여 만든다.

그 이후에 더해서 만들 수 있는 최대값을 이용해 prime수를 판별하는 배열을 만들어

prime수일 경우는 1을 아닐경우는 0을 배열에 넣어 combinations으로 만든 수를 prime배열에 대응시켜

값을 더한다.
