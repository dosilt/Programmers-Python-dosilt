```python
def solution(n):
    temp = [False, False] + [True for i in range(n-1)]
    for i in range(2, n//2+1):
        if temp[i] == True:
            for j in range(i*2, n+1, i):
                temp[j] = False
    return sum(temp)
```

```python
print(solution(10)) # 4
print(solution(5)) # 3
```

# 해설
에라토스테네스의 체를 이용하여 원래는 sqrt(n) 까지만 연산하면 되지만 기본 함수로 sqrt함수를 제공하지 않기 때문에
//2+1까지 진행하였다.
