```python
def solution(num):
    if num == 1:
        return 0

    for i in range(501):
        if num % 2 == 0: # 짝수일 경우
            num = num // 2
        else: # 홀수일 경우
            num = num*3 + 1
        if num == 1:
            return i+1
    return -1
```

```python
print(solution(6)) # 8
print(solution(16)) # 4
print(solution(626331)) # -1
```

# 해설
1-1. 입력된 수가 짝수라면 2로 나눕니다.

1-2. 입력된 수가 홀수라면 3을 곱하고 1을 더합니다.

2. 결과로 나온 수에 같은 작업을 1이 될 때까지 반복합니다.

단, 작업을 500번을 반복해도 1이 되지 않는다면 –1을 반환해 주세요.

입력 변수 n 은 1이상 8000000 미만 정수이기 때문에 0이 올수는 없지만 1이 올수도 있는 문제이므로

1일 경우에는 처리과정 없이 바로 1이기 때문에 0을 return 해 준다

위의 조건만 잘 따라서 코드로 구현할수만 있으면 큰 문제는 없는 문제다.
