```python
def solution(n):
    answer = [1, 2] + [0] * (n-2)
    for i in range(2, n):
        answer[i] = (answer[i-1] + answer[i-2]) % 1000000007
    return answer[-1]
```

```python
print(solution(4)) # 5
```

# 해설
세로의 길이는 2로 고정이고 가로의 길이가 n 일때 2 * n 의 면적을 타일로 채우려고 할 때

타일은 2 * 1 로 크기가 고정이고 총 몇가지 방법으로 채우느냐 인데 이것은 dp를 이용하면 된다

n = 1 -> 1가지

n = 2 -> 2가지

n = 3 -> 3가지

n = 4 -> 5가지

느낌이 오듯이 fibonacci랑 같다.

특이한 것은  % 1000000007 를 for문에 넣어야 효율성 테스트를 통과할 수 있다.
