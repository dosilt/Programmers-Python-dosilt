```python
def flip(arr):
    temp = arr + [0] + [0 if arr[i] == 1 else 1 for i in range(len(arr)-1, -1, -1)]
    return temp

def solution(n):
    temp = [[0]]
    for i in range(1, n):
        temp.append(flip(temp[i-1]))
    return temp[n-1]
```

```python
print(solution(1)) # [0]
print(solution(2)) # [0,0,1]
print(solution(3)) # [0,0,1,0,0,1,1]
print(solution(4))
```

# 해설
종이를 반으로 접게 되면 가운데 부분은 움푹 패인 [0] 의 값을 가지게 되며 이것을 지속적으로 진행하였을 경우

생기는 공식이 전에 접었던 종이의 모양의 배열을 A 라고 한다면 A + [0] + not(reversed(A)) (A의 배열의 순서가

반대이며 0 -> 1로 1 -> 0 으로 변경한 것을 더한 배열이 그 이후에 나오게 된다.

즉 위의 공식을 이용하여 코딩을 짜면 간단하게 끝낼 수 있다
