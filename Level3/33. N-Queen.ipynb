```python
def check(x):
    for i in range(x):
        if result[i] == result[x] or abs(result[x]-result[i]) == x-i:
            return False
    return True

def dfs(x, n):
    count = 0
    if x == n:
        return 1
    else:
        for i in range(n):
            result[x] = i
            if check(x):
                count += dfs(x+1, n)
        return count

def solution(n):
    global result
    result = [0] * n
    return dfs(0, n)
```

```python
print(solution(1)) # 1
print(solution(2)) # 0
print(solution(3)) # 0
print(solution(4)) # 2
print(solution(5)) # 10
print(solution(6)) # 4
print(solution(7)) # 40
print(solution(8)) # 92
print(solution(9)) # 352
```
