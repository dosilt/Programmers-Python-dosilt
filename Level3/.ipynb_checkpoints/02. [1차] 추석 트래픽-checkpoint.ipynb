```python
def time2sec(time):
    t = list(map(float, time.split(':')))
    return 3600*t[0] + 60*t[1] + t[2]
    
def solution(lines):
    temp = [s.split(' ')[1:] for s in lines if 0.001 <= float(s.split(' ')[2][:-1]) <= 3]
    start_list = []
    end_list = []
    for time, play in temp:
        end = time2sec(time)
        start = round(end - float(play[:-1]) + 0.001, 4)
        start_list.append(start)
        end_list.append(end)
    
    maxi = 0
    for i in range(len(end_list)):
        answer = 1
        for j in range(i+1, len(start_list)):
            if end_list[i] + 1 > start_list[j]:
                answer += 1
        if maxi < answer:
            maxi = answer
    return maxi
```

```python
print(solution(["2016-09-15 00:00:00.000 3s"])) # 1
print(solution(["2016-09-15 23:59:59.999 0.001s"])) # 1
print(solution(["2016-09-15 01:00:04.001 2.0s", "2016-09-15 01:00:07.000 2s"])) # 1
print(solution(["2016-09-15 01:00:04.002 2.0s", "2016-09-15 01:00:07.000 2s"])) # 2
print(solution(["2016-09-15 20:59:57.421 0.351s", "2016-09-15 20:59:58.233 1.181s", "2016-09-15 20:59:58.299 0.8s", "2016-09-15 20:59:58.688 1.041s", "2016-09-15 20:59:59.591 1.412s", "2016-09-15 21:00:00.464 1.466s", "2016-09-15 21:00:00.741 1.581s", "2016-09-15 21:00:00.748 2.31s", "2016-09-15 21:00:00.966 0.381s", "2016-09-15 21:00:02.066 2.62s"]))
# 7
print(solution(["2016-09-15 00:00:00.000 2.3s", "2016-09-15 23:59:59.999 0.1s"])) # 1
```

# 해설
위에 주어진 자료는 9월 15일 하루동안에 데이터 이기 때문에 그 이외에 날짜에 대해서 고민하지 않아도 되며

종료 시간 기준으로 정렬 되어 있기때문에 end time은 오름차순으로 정렬되어 있다.

따라서 시작 시간과 종료시간 list를 나눈후 종료시간+1 > 시작시간 배열을 비교해서 최대값을 구하면된다.
