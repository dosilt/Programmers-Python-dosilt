```python
def solution(s):
    arr = s.split(' ')
    answer = ''
    for i in arr:
        temp = []
        for j, data in enumerate(i):
            if j % 2 == 0: # 짝수
                temp.append(data.upper())
            else:
                temp.append(data.lower())
        answer += ''.join(temp) + ' '
    return answer[:-1]
```

```python
print(solution('try hello world')) # TrY HeLlO WoRlD
```

# 해설
각 단어에 홀수번 째 알파벳은 소문자로 짝수번 째 알파벳은 대문자로 변경을 해야된다. 따라서 split을 이용하여 각 단어를 나누고
각 단어의 짝수 번은 대문자로 홀수는 소문자로 만든후에 단어를 재조합 하고 ' ' (공백) 을 더해서 answer 이라는 문자열에 추가시킨다
끝까지 추가 시킨후에는 마지막에 ' '(공백) 이 존재하기 때문에 [:-1] 까지만 answer에서 return 한다.
