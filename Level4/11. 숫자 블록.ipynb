```python
def solution(begin, end):
    l = [0] * (end+1)
    for i in range(end//2, begin-1, -1):
        for j in range(i*2, end+1, i):
            if l[j] == 0:
                l[j] = i
    return l[1:]
```

```python
print(solution(1, 10)) # [0, 1, 1, 2, 1, 3, 1, 4, 3, 5]
```

```python
from math import sqrt
 
def solution(begin, end):
    result=[]
    if begin==1:
        result.append(0)
        begin+=1
    for i in range(begin,end+1):
        for j in range(2,int(sqrt(end)+1)):            
            if i%j==0:
                result.append(i//j)
                break
        else : result.append(1)
 
    return result
```
